# Quantitative finance: Analysis of Kames Ethical Equity Fund

We found that this fund is a good alternative to get exposure to the UK stock index (the FTSE All Share TR GBP represent 98-99% of UK market capitalisation) while incorporating environmental, social and governance (ESG) factors into investment decisions.<br>
Kames Ethical Equity Fund performances and risk metrics are very close to his benchmark.

## 1. Description

Kames Capital is a UK-based investment manager with £42.6 billions AUM$^{1}$ subsidiary of Aegon NV Group, a life insurance, pensions and asset management company headquartered in Netherlands.

Kames Ethical Equity Fund describes himself as an ethical fund focused on equity and equity type securities in companies either based in the UK, listed in the UK stock market or conducting business in the UK meeting the funds predefined ethical criterions.$^{2}$ It’s objective is to maximise total return while complying with this strategy. The fund benchmark is the FTSE All Share TR GBP.$^{3}$ 

The ethical policy is to avoid investing in companies that it decides harm people, society,
animals or the environment. An external company$^{4}$ provides an ethical screening based on these ethical criterions. To sum up these criterions by each letter of the acronym ESG:$^{5}$
- Environment: excludes companies that "are involved in activities which have a significant negative impact in the environment ..., who have been convicted of serious pollution offences ... or companies in energy intensive industries which are not tackling the issue of climate change"
- Social: excludes big beverage companies, all pornography companies, gambling companies and providers of slaughterhouse by-products or fish
- Governance: excludes companies that are providing nuclear power, military companies and companies operating in countries with poor Human Rights records **without established management policies on these issues** (people need to make end meet).

The sector allocation$^{6}$ as of December 31st, 2018 point out the impact of these criterions with a strong focus on services industry – representing around 60% of the NAV. In the meantime utilities sector, Oil & Gas sector and Basic Materials sectors weights around 10%. The ethical criterion on animal welfare excludes health care companies that are massive users of animals for clinical trials.

### 2.3. Performance measures

The Sharpe ratio is a classification criterion that determine which portfolio is the most efficient for the risk taken.

```python
print("Sharpe ratios:\nKames Ethical Equity Fund:{0:8.4f} \nFTSE All Share TR GBP (Benchmark):{1:8.4f}".format(
    sharpeFund, sharpeBench))
```

Invest in the fund is a better than buying the benchmark on a risk-adjusted basis.<br>
This ratio is relevant when the distribution of returns follows a Normal distribution (as we only consider the mean and the standard deviation), therefore non-symmetric distributions are not accurately represented. This metric is still useful as our portfolio is not diversified (UK-focus equity fund).

The Sortino ratio is another classification criterion that determine which portfolio is the most efficient for the risk taken. The theory behind the Sortino ratio is that the riskiness of a portfolio is better measured by the deviation of returns below a target return. That is why this ratio is better suited to evaluate high-volatility portfolios.

```python
print("Sortino ratios:\nKames Ethical Equity Fund:{0:8.4f} \nFTSE All Share TR GBP (Benchmark):{1:8.4f}".format(
    sortinoFund, sortinoBench))
```

Once again the fund perform better than his benchmark in terms of downside risk.

One more classification criterion: Omega ratio. Unlike Sharpe ratio that only considers the mean and the standard deviation, this metric condiders all moments. Also Omega ratio includes include higher levels of information specific to the shape of the distribution of the underlying return data.

```python
print("Omega ratios:\nKames Ethical Equity Fund:{0:8.4f} \nFTSE All Share TR GBP (Benchmark):{1:8.4f}".format(
    omegaFund, omegaBench))
```

We see that there is a much more small difference between the fund and his benchmark, as they have the same chance as far as earning a target return equal to SONIA.

Jensen's alpha is a risk-adjusted performance benchmark that tells you how by much the returns of an actively managed portfolio are above or below market returns.

```python
print("Jensen's Alpha: {0:8.4f}%".format(alphaJensen * 100))
```

This very little alpha was suggest by the results that we have with Omega ratio.

The appraisal ratio measures the performance of a portfolio manager for each point of volatility granted to its benchmark.<br>
For funds with similar values of Jensen’s Alpha, a higher Information Ratio indicates a better managed fund with superior stock picking. However, this is only valid if the fund and its benchmark are strongly correlated.

```python
print("Appraisal ratio: {0:8.4f}%".format(IR * 100))
```

## 3. Qualitative analysis

Our in-depth analysis of the fund strategy gave us some hints to choose our factors:
- the fund is focusing on UK stock market: we need to have macroeconomic factors related to the UK economy.
- the fund is investing both in small and big cap stocks : we should consider style factor.

We selected 15 factor to conduct our study:
- macroeconomic factors:
    - UK Inflation$^{7}$
    - UK Employment$^{8}$
    - UK Household consumption$^{9}$
    - UK GDP Index$^{10}$
    - UK Business confidence index$^{11}$
    - GBP EUR exchange rate$^{12}$
    - GBP USD exchange rate$^{13}$
    - Bank of England base rate$^{14}$
    - UK retail sales$^{15}$
    - UK consumer confidence$^{16}$
    - UK balance of payments$^{17}$
- style factors:
    - Small Minus Big, a factor measuring the historic excess of small-cap companies over big-cap companies. It is a size factor.
    - High Minus Low, a factor representing the spread in returns between companies with a high book-to-market ratio (value companies) and companies with a low book-to-market ratio (growth companies). It is a value factor.
    - Up Minus Down, a factor that measures the (historical) excess returns of the winners that went up minus the losers that lost value. It is a momemtum factor.
    These factors have been calculated and provided by researchers from University of Exeter$^{18}$.
- ESG factor: MSCI UK Leaders Index$^{19}$

###  1. The data

All data are from Bloomberg if nothing is mentioned on the origin.

```python
def excelToDf(theTable):
    cols = next(theTable)[1:]
    theTable = list(theTable)
    idx = [r[0] for r in theTable]
    theTable = (islice(r, 1, None) for r in theTable)
    theTable = pd.DataFrame(theTable, index = idx, columns = cols)
    theTable.index = theTable.index.map(pd.DatetimeIndex)
    return theTable
```

```python
theData = openpyxl.load_workbook(filename = "Data_Partie_1.xlsx")
dataFund = excelToDf(theData["NAV"].values)
SONIA = excelToDf(theData["SONIA"].values)
benchPrice = excelToDf(theData["FTSE All Share TR"].values)
macroFactors = excelToDf(theData["Macro Factors"].values)
styleFactors = excelToDf(theData["Style Factors (UK) Monthly"].values)
esgFactors = excelToDf(theData["ESG Factors"].values)
styleFactorsD = excelToDf(theData["Style Factors (UK) Daily"].values)
```

### 2. Performance summary calculation

Returns

```python
startD = dataFund.index.get_loc("2013-11-29") + 1
dataFund5y = dataFund.iloc[startD:,::]
```

```python
dataFund["Daily_return"] = dataFund[["NAV in GBP"]] / dataFund[["NAV in GBP"]].shift(1) - 1
dataFund = dataFund.merge(SONIA, how = "left", left_index = True, right_index = True)
dataFund = dataFund.replace("", np.nan, regex = True).dropna()
dataFund["Excess_return"] = dataFund["Daily_return"] - dataFund["SONIA daily return"]
avgDailyReturn = dataFund["Daily_return"].mean()
avgDailyXsReturn = dataFund["Excess_return"].mean()
```

```python
dataFund5y["Daily_return"] = dataFund5y[["NAV in GBP"]] / dataFund5y[["NAV in GBP"]].shift(1) - 1
dataFund5y = dataFund5y.merge(SONIA, how = "left", left_index = True, right_index = True)
dataFund5y["Excess_return"] = dataFund5y["Daily_return"] - dataFund5y["SONIA daily return"]
dataFund5y = dataFund5y.replace("", np.nan, regex = True).dropna()
avgDailyReturn5y = dataFund5y["Daily_return"].mean()
avgDailyXsReturn5y = dataFund5y["Excess_return"].mean()
```

Cumulative returns

```python
N = len(dataFund.index.unique())
dataFund["Cumulative_Return"] = dataFund["Daily_return"].add(1).cumprod().add(-1)
dataFund["Cumulative_XS_Return"] = dataFund["Excess_return"].add(1).cumprod().add(-1)
return2006 = float(dataFund[["Cumulative_Return"]].tail(1).values)
XSreturn2006 = float(dataFund[["Cumulative_XS_Return"]].tail(1).values)
returnY = ((1 + dataFund.loc["2018-11-30","Cumulative_Return"]) ** (1/(N/252))) - 1
XSreturnY = ((1 + dataFund.loc["2018-11-30","Cumulative_XS_Return"]) ** (1/(N/252))) - 1
```

```python
N2 = len(dataFund5y.index.unique())
dataFund5y["Cumulative_Return"] = dataFund5y["Daily_return"].add(1).cumprod().add(-1)
dataFund5y["Cumulative_XS_Return"] = dataFund5y["Excess_return"].add(1).cumprod().add(-1)
return5y = float(dataFund5y[["Cumulative_Return"]].tail(1).values)
XSreturn5y = float(dataFund5y[["Cumulative_XS_Return"]].tail(1).values)
returnY5y = ((1 + dataFund5y.loc["2018-11-30","Cumulative_Return"]) ** (1/(N2/252))) - 1
XSreturnY5y = ((1 + dataFund5y.loc["2018-11-30","Cumulative_XS_Return"]) ** (1/(N2/252))) - 1
```

```python
summaryReturn = pd.DataFrame({"Return" : {"Average daily return" : avgDailyReturn * 100, 
                    "Cumulative return" : return2006 * 100, "Cumulative Annualised return" : returnY * 100},
                    "Excess return" : {"Average daily return" : avgDailyXsReturn * 100 , 
             "Cumulative return" : XSreturn2006 * 100, "Cumulative Annualised return" : XSreturnY * 100}})
```

```python
summaryReturn5y = pd.DataFrame({"Return" : {"Average daily return" : avgDailyReturn5y * 100, 
                    "Cumulative return" : return5y * 100, "Cumulative Annualised return" : returnY5y * 100},
                    "Excess return" : {"Average daily return" : avgDailyXsReturn5y * 100 , 
             "Cumulative return" : XSreturn5y * 100, "Cumulative Annualised return" : XSreturnY5y * 100}})
```

Volatility

```python
dailyVol = dataFund["Daily_return"].std()
volY = dailyVol * np.sqrt(252)

benchPrice = benchPrice.replace("", np.nan, regex = True).dropna()
benchPrice["Daily_return_B"] = benchPrice[["Benchmark"]] / benchPrice[["Benchmark"]].shift(1) - 1
dailyVolBench = benchPrice["Daily_return_B"].std()
volYBench = dailyVolBench * np.sqrt(252)
```

Numbers of days

```python
N = dataFund.shape[0]
```

Drawdown

```python
maxPerf = dataFund["Cumulative_Return"].cummax()
dataFund["Drawdown"] = dataFund["Cumulative_Return"].subtract(maxPerf)
```

Tracking Error

```python
dataFund = dataFund.merge(benchPrice, how = "left", left_index = True, right_index = True)
dataFund["(Fund - Bench)^2"] = (dataFund["Daily_return"] - dataFund["Daily_return_B"]) ** 2
TE = np.sqrt(dataFund["(Fund - Bench)^2"].sum() / (N - 1))
```

Sharpe ratio

```python
sharpeFund = (((1 + dataFund["Excess_return"].add(1).cumprod()[-1]) ** (1/(N/252))) - 1) / volY

dataFund["Excess_return_B"] = dataFund["Daily_return_B"] - dataFund["SONIA daily return"]
sharpeBench = (((1 + dataFund["Excess_return_B"].add(1).cumprod()[-1]) ** (1/(N/252))) - 1) / volY
```

Sortino ratio

```python
negativeVol = dataFund[dataFund["Excess_return"] < 0]["Excess_return"].std() * np.sqrt(252)
negativeVolBench = dataFund[dataFund["Excess_return_B"] < 0]["Excess_return_B"].std() * np.sqrt(252)

sortinoFund = (((1 + dataFund["Excess_return"].add(1).cumprod()[-1]) ** (1/(N/252))) - 1) / negativeVol
sortinoBench = (((1 + dataFund["Excess_return_B"].add(1).cumprod()[-1]) ** (1/(N/252))) - 1) / negativeVolBench
```

Jensen's Alpha

```python
Ep = dataFund["Daily_return"].mean() ; Em = dataFund["Daily_return_B"].mean()
rf = dataFund["SONIA daily return"].mean()
cova = dataFund[["Daily_return", "Daily_return_B"]].cov().iloc[0,1]
beta = cova / np.var(dataFund["Daily_return_B"])
alphaJensen = Ep - (rf + beta * (Em - rf))
```

Appraisal ratio

```python
IR = np.mean(dataFund["Daily_return"] - dataFund["Daily_return_B"]) / TE
```

Omega ratio

```python
posR = dataFund[dataFund["Daily_return"] > dataFund["SONIA daily return"]][["Daily_return", "SONIA daily return"]]
posR["Diff"] = posR["Daily_return"] - posR["SONIA daily return"]
negR = dataFund[dataFund["Daily_return"] < dataFund["SONIA daily return"]][["Daily_return", "SONIA daily return"]]
negR["Diff"] = negR["Daily_return"] - negR["SONIA daily return"]
omegaFund = posR["Diff"].sum() / (- negR["Diff"].sum())

posRBench = dataFund[dataFund["Daily_return_B"] > dataFund["SONIA daily return"]][["SONIA daily return", "Daily_return_B"]]
posRBench["Diff"] = posRBench["Daily_return_B"] - posRBench["SONIA daily return"]
negRBench = dataFund[dataFund["Daily_return_B"] < dataFund["SONIA daily return"]][["SONIA daily return", "Daily_return_B"]]
negRBench["Diff"] = negRBench["Daily_return_B"] - negRBench["SONIA daily return"]
omegaBench = posRBench["Diff"].sum() / (- negRBench["Diff"].sum())
```
